```python
tumor_subset = tumor_adata[tumor_adata.obs['leiden_res_0.10'].isin(['0','2'])]
# Define a function to map values from ColumnA to ColumnB
def map_values(value):
    if value == '0':
        return 'classical'
    else:
        return 'basal'

tumor_subset.obs['pdac_subtype'] = tumor_subset.obs['leiden_res_0.10'].apply(map_values)

sc.pl.umap(tumor_subset, color=['pdac_subtype'], 
           legend_loc="on data",
           frameon=False)
display(tumor_subset.X)


def anndata_to_celltype_df(adata, celltype_column='pdac_subtype'):
    # Extract gene expression matrix
    expression_df = pd.DataFrame(adata.X.T, index=adata.var_names, columns=adata.obs_names)
    
    # Get cell type information
    celltypes = adata.obs[celltype_column]
    
    # Aggregate data by cell type
    grouped = expression_df.groupby(celltypes, axis=1)
    aggregated_df = grouped.mean()
    
    return aggregated_df
```

```python
reference = anndata_to_celltype_df(tumor_subset)
bulk_metadata = pd.read_csv('ck_primitive_annotation.csv')
bulk_counts = pd.read_csv('ck_primitive_normalized_counts.csv',index_col = 'Unnamed: 0')
```
